We have the min and max temperatures in a city In India for each months of the year. We
would like to find a function to describe this and show it graphically, the dataset given below.
Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

temp_months = np.arange(12)
plt.plot(temp_months, temp_max, 'ro')
plt.plot(temp_months, temp_min, 'bo')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
```

fitting it to the periodic function

```python
from scipy import optimize
def temp_year(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(temp_year, temp_months, temp_max, [40, 10, 0])
res_min, cov_min = optimize.curve_fit(temp_year, temp_months, temp_min, [20, 10, 0])
```

```python

days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(temp_months, temp_max, 'ro')
plt.plot(days, temp_year(days, *res_max), 'r-')
plt.plot(temp_months, temp_min, 'bo')
plt.plot(days, temp_year(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')

plt.show()
```
